## Get data
The first thing we need to do in order to build our trade is to get the data. One method is to go to Quandl and pay for an API, but in true "hack3r" form, why pay for data when you can take a few steps to extract it for free? 

Therefore, let's go get our investing universe, the S&P 500 and go about the process of building a data frame. 

```python
end_date = datetime.now()
start_date = end_date + timedelta(-365*5)
```

```python
sp500_df = client.get_dataframe(sp500,
                                metric_name='close',
                                startDate=start_date,
                                endDate=end_date,
                                frequency='daily'
                               )
```
