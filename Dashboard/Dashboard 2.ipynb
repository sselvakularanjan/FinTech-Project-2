```python
# Load the data
sp500_ticker_df = pd.read_csv('Resources/sp500_close.csv', index_col=0, parse_dates=True)
sp500_ticker_df.head()
```

```python
dcc.Graph(id='timeseries',
          config={'displayModeBar': False},
          animate=True,
          figure=px.line(sp500_ticker_df,
                         x='Date',
                         y='value',
                         color='stock',
                         template='plotly_dark').update_layout(
                                   {'plot_bgcolor': 'rgba(0, 0, 0, 0)',
                                    'paper_bgcolor': 'rgba(0, 0, 0, 0)'})
                                    )
```
