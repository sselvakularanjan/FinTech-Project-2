```python
# Import Libraries and Dependencies 
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import urllib.request
import matplotlib.pyplot as plt
import os
import json
from pathlib import Path
from datetime import datetime, timedelta
from tiingo import TiingoClient
```

```python
# Bring Env Variables into Python/Store in Variable (go to keys.sh file and activate)
tiingo = os.getenv('TIINGO_API_KEY')
```

```python
# Create a Client to Connect to Tiingo
client = TiingoClient()
```

```python
# for dynamic date windows back one year
end = datetime.now()
start = end + timedelta(-365*5)
```

```python
mom_mid_df = pd.read_csv("Resources/sp500_mom_mid.csv")
mom_mid_df = mom_mid_df.drop(columns="date", index=mom_mid_df.index)
```

```python
mom_short_df = pd.read_csv("Resources/sp500_mom_short.csv")
mom_short_df = mom_short_df.drop(columns="date", index=mom_mid_df.index)
```

```python
mom_mid_tickers=[]
mom_mid_tickers=mom_mid_df.columns
mom_mid_tickers
```

```python
mom_short_tickers=[]
mom_short_tickers=mom_short_df.columns
mom_short_tickers
```

```python
# Create DataFrame for SP500 constituents' close price (takes a few minutes for 1 year's data)
mom_tickers_mid_df = client.get_dataframe(mom_mid_tickers,
                                            metric_name='close',
                                            startDate= start,
                                            endDate= end,
                                            frequency='daily')
mom_tickers_mid_df.head()
```

```python
#create a CSV file to save the data
mom_tickers_mid_df.to_csv('Resources/mom_tickers_mid.csv',index=['date'])
```

```python
# Create DataFrame for SP500 constituents' close price (takes a few minutes for 1 year's data)
mom_tickers_short_df = client.get_dataframe(mom_short_tickers,
                                            metric_name='close',
                                            startDate= start,
                                            endDate= end,
                                            frequency='daily')
mom_tickers_short_df.head()
```

```python
#create a CSV file to save the data
mom_tickers_short_df.to_csv('Resources/mom_tickers_short.csv',index=['date'])
```
